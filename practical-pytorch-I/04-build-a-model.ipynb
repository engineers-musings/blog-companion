# Practical PyTorch · Part 4 — Companion Notebook

Build a model with your own hands: a single `nn.Linear`, then a small stack, then count and name every parameter inside. Nothing here is trained — we only assemble and look.

## 1. The simplest model: a single nn.Linear

`nn.Linear` does a weighted sum: numbers in, multiply by learned weights, add a bias, numbers out. It's already a complete `nn.Module`. Inside it live two parameters — a `weight` and a `bias`.

```python
import torch
import torch.nn as nn

layer = nn.Linear(3, 2)   # 3 numbers in, 2 numbers out
print(layer)              # Linear(in_features=3, out_features=2, bias=True)

print(layer.weight.shape)   # torch.Size([2, 3])
print(layer.bias.shape)     # torch.Size([2])

x = torch.rand(1, 3)   # one example, 3 features
out = layer(x)
print(out.shape)       # torch.Size([1, 2]) — a forward pass
```

## 2. Stack a few: a tiny network

`nn.Sequential` runs layers in order, top to bottom. The `nn.ReLU` in the middle keeps the positives (negatives become zero) — the nonlinear kink that makes depth count.

```python
net = nn.Sequential(
    nn.Linear(10, 5),   # weighted sum: 10 numbers in, 5 out
    nn.ReLU(),          # keep the positives
    nn.Linear(5, 2),    # weighted sum: 5 numbers in, 2 out
)
print(net)

x = torch.rand(1, 10)   # one example, 10 features
out = net(x)
print('output shape:', tuple(out.shape))   # (1, 2) — data flowed through the stack
```

## 3. Counting and naming the parameters

The parameters are the actual numbers inside the layers. Count them all in one line, then walk them by name — the *same* code you'll later run on a model with millions.

```python
total = sum(p.numel() for p in net.parameters())
print(f'{total} parameters')   # 67 parameters

for name, p in net.named_parameters():
    print(name, tuple(p.shape))
# 0.weight (5, 10) / 0.bias (5,) / 2.weight (2, 5) / 2.bias (2,)
# '0' is the first Linear, '2' the second — index 1 is the ReLU, which has no numbers
```

Four tensors, sixty-seven numbers, no secrets — and the same `named_parameters()` walk works on an eleven-million-parameter network.
