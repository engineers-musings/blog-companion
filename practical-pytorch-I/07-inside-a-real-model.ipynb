# Practical PyTorch · Part 7 — Companion Notebook

Open a real, pretrained deep network — ResNet-18 — and read it with the exact skills you built on a toy: the layer tree, eleven million parameters, and the named weights. We only look — no training.

## 1. The three layers you'll see everywhere

- **nn.Linear** — a weighted sum (Part 4). The last one turns internals into one score per category.
- **nn.Conv2d** — slides a small filter over an image (Part 6, the heart of LeNet). The vision workhorse.
- **nn.ReLU** — keep the positives.

You'll also spot `BatchNorm2d`, `MaxPool2d`, and `Dropout` — supporting cast you can read past on a first pass.

## 2. Load ResNet-18 and read it

`weights=ResNet18_Weights.DEFAULT` asks for the best official pretrained weights (downloaded once, then cached). Then do the same thing you did with your toy — just `print` it and read the shape of the tree.

```python
from torchvision.models import resnet18, ResNet18_Weights

model = resnet18(weights=ResNet18_Weights.DEFAULT)
print(model)
```

The indented outline *is* the tree — the same kind your tiny net had, just deeper. Spot the familiar faces: `Conv2d`, `ReLU`, `Sequential`, and a closing `Linear` (`fc`) producing one score per ImageNet category.

## 3. Count the parameters

The *same* counting line you ran on your three-layer net — there are simply more of them. Then peek at where the first few live; the names mirror the tree.

```python
total = sum(p.numel() for p in model.parameters())
print(f'{total:,} parameters')   # 11,689,512 parameters  (your toy had 67)

for name, p in list(model.named_parameters())[:5]:
    print(name, tuple(p.shape))
# conv1.weight (64, 3, 7, 7) / bn1.weight (64,) / ... / layer1.0.conv1.weight (64, 64, 3, 3)
```

Every parameter is a tensor with a shape — exactly like the `weight` and `bias` on your single `Linear`, just more of them. The box is glass, even at eleven million numbers.
