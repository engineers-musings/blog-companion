## 1. The easy button: pipeline

```python
from transformers import pipeline

gen = pipeline("text-generation", model="distilgpt2")
out = gen("The best thing about PyTorch is", max_new_tokens=30)
print(out[0]["generated_text"])
```

## 2. One level down: the autoregressive loop

`AutoModelForCausalLM` is the generation task class; `model.generate` runs the loop (predict → append → repeat).

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tok = AutoTokenizer.from_pretrained("distilgpt2")
model = AutoModelForCausalLM.from_pretrained("distilgpt2")

inputs = tok("The best thing about PyTorch is", return_tensors="pt")
output_ids = model.generate(**inputs, max_new_tokens=30)
print(tok.decode(output_ids[0], skip_special_tokens=True))
```

GPT-2 *continues* text — it doesn't follow instructions. Ask it a question and it tends to wander. Decoding controls (Part 24) and instruction-tuned models (Part 21) fix that.
