Moving tensors between the CPU and GPU with `.to(device)`, trading data with NumPy, and pulling a single value out with `.item()`. Everything runs on CPU too — a GPU just makes the big ones faster.

## 1. Is a GPU even there?

`cuda` is NVIDIA's GPU technology — what Colab gives you under **Runtime → Change runtime type → GPU**. If this prints `False`, everything below still runs; it just won't get the speedup.

```python
import torch

print(torch.cuda.is_available())   # True if a GPU is attached, else False
```

## 2. The one device line, and `.to(device)`

Pick the best available device once, into a variable — then the same notebook runs untouched with or without a GPU.

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print(device)

x = torch.rand(2, 3)
print(x.device)        # cpu — where it was born

x = x.to(device)
print(x.device)        # cuda (if attached) or cpu
```

A CPU tensor and a GPU tensor behave **identically** — same shape, indexing, and math. The only difference is speed, and it only shows up once tensors get large. Move the inputs, and the result simply appears on the same device.

```python
a = torch.rand(1000, 1000).to(device)
b = torch.rand(1000, 1000).to(device)

c = a @ b              # runs on the GPU if one's attached
print(c.device)        # the result lands wherever its inputs were
```

## 3. Talking to NumPy

The world outside PyTorch speaks NumPy, which only understands CPU data. Bring a tensor home with `.cpu()` before converting. Note: `from_numpy` / `.numpy()` *share memory* — `.clone()` if you want an independent copy.

```python
import numpy as np

t = torch.tensor([1.0, 2.0, 3.0])
arr = t.numpy()                 # tensor -> NumPy array (CPU only)
print(arr)

arr2 = np.array([4.0, 5.0, 6.0])
back = torch.from_numpy(arr2)   # NumPy array -> tensor
print(back)

result = x.cpu().numpy()        # GPU -> CPU -> NumPy
print(result.shape)
```

## 4. Pulling out a single value with `.item()`

A reduction like `.mean()` hands back a tensor holding *one* number. `.item()` unwraps it as a plain Python `float` or `int` — the move when you want to print it, compare it, or store it.

```python
score = torch.tensor(0.98)
print(score.item())             # 0.98 — a normal Python float

g = torch.tensor([1.0, 2.0, 3.0])
print(g.mean().item())          # 2.0 — reduce, then unwrap
```
