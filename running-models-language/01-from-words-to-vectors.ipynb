**From Words to Vectors: How a Model Turns Text into Numbers.**

A model only does arithmetic — it cannot multiply the word "cat." So before the transformer can let words look at each other, each word has to become numbers. Two moves: give every known word an ID, then turn that ID into a vector. We build both on a twelve-word toy vocabulary. No training.

## 1. A vocabulary is just a numbered list

A model knows a fixed list of words — its **vocabulary** — and each word's position in that list is its **ID**.

```python
vocab = ["the", "a", "cat", "dog", "sat", "ran",
         "on", "mat", "sofa", "and", "big", "small"]

stoi = {word: i for i, word in enumerate(vocab)}
print(stoi["cat"])   # 2
```

## 2. A sentence becomes a row of IDs

Split the sentence into words and swap each for its ID. The same word always maps to the same ID.

```python
sentence = "the cat sat on the mat"
ids = [stoi[word] for word in sentence.split()]
print(ids)   # [0, 2, 4, 6, 0, 7]
```

Two cracks the real tokenizers in Part 16 seal: an unknown word like `"kitten"` raises `KeyError`, and splitting on spaces ignores punctuation and casing. Real tokenizers break text into *subword* pieces so they can always compose a word they have never seen.

## 3. IDs become vectors — the embedding lookup

IDs are arbitrary labels: `cat` being `2` does not make it "less" than `sofa` at `8`. We want a vector per word instead. `nn.Embedding` is a table with one row per word; looking up an ID grabs that row.

```python
import torch
import torch.nn as nn

embed = nn.Embedding(num_embeddings=len(vocab), embedding_dim=4)
print(embed.weight.shape)   # torch.Size([12, 4])
```

```python
x = torch.tensor(ids)        # [0, 2, 4, 6, 0, 7]
vectors = embed(x)
print(vectors.shape)         # torch.Size([6, 4])

# the two 0s ("the") come back as the same row
print(torch.equal(vectors[0], vectors[4]))   # True
```

The table starts **random**, so right now `cat` and `dog` are no closer than `cat` and `sofa`. The meaning is *learned* during training, until words that behave alike drift together. **Part 20** is all about reading that geometry once it exists.
