## Load a writer

The model downloads the first time you run it. `distilgpt2` is a small, fast writing model — perfect for experimenting.

```python
!pip install -q transformers
from transformers import pipeline

writer = pipeline("text-generation", model="distilgpt2")
```

## Give it a starting line

The AI takes your opening and continues it, inventing the rest. Run it a few times — you'll get a different story each time. The result is a **list with a dictionary**; `result[0]["generated_text"]` pulls out the text.

```python
result = writer("Once upon a time, a brave little robot", max_length=40)
print(result[0]["generated_text"])
```

### Try it 🎯

1. Change the starting line to your own — a story opener, a joke setup, a recipe.
2. Change `max_length=40` to `60` for a longer piece.
3. Run the same prompt three times. Notice it's different each time.
