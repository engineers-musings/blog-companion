Load the data and chart tool first.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
penguins = sns.load_dataset("penguins").dropna()
```

## A title and labels

Add words with `plt.title`, `plt.xlabel`, and `plt.ylabel` — *before* `plt.show()`.

```python
penguins.groupby("species")["body_mass_g"].mean().plot(kind="bar")
plt.title("Average penguin weight by species")
plt.xlabel("Species")
plt.ylabel("Body mass (grams)")
plt.show()
```

## Add some color

`.plot(...)` accepts a `color`. Try `"coral"`, `"slateblue"`, `"gold"` — pick something that fits your story.

```python
penguins["species"].value_counts().plot(kind="bar", color="teal")
plt.title("How many penguins of each species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

## Make it bigger

Set the size with `figsize=(width, height)` (in inches). Notice the *title itself* states the finding.

```python
penguins.plot(kind="scatter", x="flipper_length_mm", y="body_mass_g", figsize=(8, 5))
plt.title("Heavier penguins have longer flippers")
plt.xlabel("Flipper length (mm)")
plt.ylabel("Body mass (grams)")
plt.show()
```

## Save your chart

Save a chart to an image file with `plt.savefig(...)` — put it just before `plt.show()`. In Colab, the file appears in the **Files** panel on the left (the folder icon); right-click it to download.

```python
penguins["species"].value_counts().plot(kind="bar", color="teal")
plt.title("Penguins by species")
plt.savefig("penguins_by_species.png")
plt.show()
```
