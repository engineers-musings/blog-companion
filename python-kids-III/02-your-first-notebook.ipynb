## A handy shortcut: the last line shows itself

In a notebook, the **last line of a cell automatically displays its value** — no `print` needed.

```python
2 + 2
```

## Cells share memory

All the cells share the same memory, so a variable you make in one cell is available in the next. Run cell 1, then cell 2.

```python
# cell 1
name = "Ada"
age = 12
```

```python
# cell 2 — run this AFTER cell 1
print(name, "is", age, "years old")
```

```python
# a list and a for-each
scores = [90, 85, 100, 70]
total = 0
for s in scores:
    total = total + s
print("Average:", total / len(scores))
```

## Fix the bug 🐞

If you run this *before* the cell that makes `total`, you get a `NameError`. The fix: run the cell that creates `total` first — like `total = 5 + 3`. In notebooks, run cells top to bottom, or use *Runtime → Run all*.

```python
total = 5 + 3
print("The total is", total)
```
