A debugging playbook for beginner inference: most errors land in one of a few buckets.

## First move: print the environment

```python
import torch
print('torch:', torch.__version__)
print('cuda available:', torch.cuda.is_available())
if torch.cuda.is_available():
    print('gpu:', torch.cuda.get_device_name(0))
```

## Read the traceback bottom-up

The **last line** of a traceback is the real error; everything above is the call chain, mostly library internals. Start at the bottom for the error type, then scan up for the last line that names *your* file — that's where you can act.

## Put model and inputs on the same device

`Expected all tensors to be on the same device` means the model and inputs disagree. Fix:

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
# model = model.to(device)
# batch = batch.to(device)
# for tokenizer dicts: inputs = {k: v.to(device) for k, v in inputs.items()}
print('device:', device)
```

## When shapes break, print shapes

`mat1 and mat2 shapes cannot be multiplied`, `expected 4D input`, `size mismatch` — label every dimension.

```python
x = torch.rand(3, 224, 224)        # missing the batch dim!
print(x.shape)                     # [3, 224, 224]
x = x.unsqueeze(0)
print(x.shape)                     # [1, 3, 224, 224] — fixed
```

## dtype mismatch

The *kind* of number has to match too — `expected scalar type Long but found Float` and friends. The fix is a cast, not a reshape: floats for model weights and image inputs; longs (`int64`) for token IDs, class indices, and labels.

```python
x = torch.rand(4)              # float32
print(x.long().dtype)          # torch.int64  — cast to integer (labels/IDs)

ids = torch.tensor([1, 2, 3])  # int64
print(ids.float().dtype)       # torch.float32 — cast to float (model input)
```

## The checklist

When something breaks, ask in order:

1. Is the package installed?
2. Is the runtime what I think it is?
3. Is CUDA available if I need it?
4. Are model and inputs on the same device?
5. Do the tensor shapes match what the model expects?
6. Did I use the model's own tokenizer / processor / transforms?
7. Is the model gated or license-restricted?
8. Is the model too large for this GPU?

For `CUDA out of memory`: restart the runtime, then smaller model, `dtype=torch.float16`, 4-bit, smaller batch, bigger GPU.
