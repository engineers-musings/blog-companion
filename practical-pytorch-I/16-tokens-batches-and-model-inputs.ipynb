How text becomes model input: token IDs, attention masks, padding, and truncation — the bridge before `pipeline()`.

## Models read token IDs, not words

```python
!pip install -q transformers
```

```python
from transformers import AutoTokenizer

name = 'distilbert-base-uncased-finetuned-sst-2-english'
tok = AutoTokenizer.from_pretrained(name)

inputs = tok('I love this movie.', return_tensors='pt')
print(inputs['input_ids'])        # integer token IDs
print(inputs.keys())              # input_ids + attention_mask
```

## See what the IDs mean

Turn the integers back into tokens to see what the tokenizer did — note the `[CLS]`/`[SEP]` special tokens it adds on every input.

```python
ids = inputs['input_ids'][0]
print(tok.convert_ids_to_tokens(ids))
# ['[CLS]', 'i', 'love', 'this', 'movie', '.', '[SEP]']
print(tok.decode(ids))   # collapse back into a readable string
```

## Tokens are not always words

Long or rare words split into sub-word pieces.

```python
print(tok.tokenize('unbelievably practical'))
```

## Batching means padding

`padding=True` pads the shorter input to the longest in the batch; the attention mask marks the filler.

```python
batch = tok(['short', 'this one is a little longer'],
            padding=True, return_tensors='pt')
print(batch['input_ids'].shape)   # [2, sequence_length]
print(batch['attention_mask'])    # 0s mark padding
```

## Truncation protects you from long inputs

```python
long_text = 'This sentence repeats. ' * 100
inputs = tok(long_text, truncation=True, max_length=512, return_tensors='pt')
print(inputs['input_ids'].shape)  # capped at [1, 512]

# You'll RUN the forward pass with these inputs by hand in Part 18.
# Next: Part 17 — The Fast Path: Hugging Face Pipelines.
```
