Load the data and chart tool first.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
penguins = sns.load_dataset("penguins").dropna()
```

## Histogram: how are the values spread?

A **histogram** sorts a number column into bins and shows how many fall in each — the *shape* of your data.

```python
penguins["body_mass_g"].plot(kind="hist", bins=20)
plt.show()
```

## Scatter plot: do two things go together?

A **scatter plot** puts one number on the bottom and another up the side, with a dot per penguin. It reveals **relationships**.

```python
penguins.plot(kind="scatter", x="flipper_length_mm", y="body_mass_g")
plt.show()
```

**Try it:** Make a scatter plot of `bill_length_mm` (x) versus `bill_depth_mm` (y).

## Line chart: a trend in order

A **line chart** connects points in order. Sorted lightest to heaviest, the line climbs steadily.

```python
top = penguins.sort_values("body_mass_g").reset_index()
top["body_mass_g"].plot(kind="line")
plt.show()
```
