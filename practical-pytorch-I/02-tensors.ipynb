A tensor is just an array of numbers that can live on a GPU. That's the whole idea — the rest is button-pressing.

## 1. Making a tensor

Hand `torch.tensor` a list (or a list of lists) and you get a tensor. The nesting depth is the number of dimensions.

```python
import torch

x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])
print(x)
```

Most of the time you ask PyTorch to build a tensor of a given shape and fill it for you. These four show up constantly.

```python
print(torch.zeros(2, 3))     # a 2x3 grid of zeros
print(torch.ones(2, 3))      # a 2x3 grid of ones
print(torch.arange(0, 10))   # 0, 1, 2, ... 9 — like Python's range()
print(torch.rand(2, 3))      # a 2x3 grid of random numbers between 0 and 1
```

## 2. Shape, dtype, device — the three questions to ask any tensor

When a tensor confuses you, print these three. They answer almost everything.

```python
x = torch.rand(2, 3)

print("shape:", x.shape)    # torch.Size([2, 3]) — how big, in each dimension
print("dtype:", x.dtype)    # torch.float32     — what kind of number
print("device:", x.device)  # cpu               — where it physically lives
```

## 3. Indexing, slicing, reshaping

Square brackets pull out pieces. The comma separates dimensions; a bare `:` means "all of this dimension."

```python
x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

print(x[0])        # first row:    tensor([1, 2, 3])
print(x[0, 1])     # row 0, col 1: tensor(2)
print(x[:, 0])     # every row, first column: tensor([1, 4])
```

Reshaping rearranges the same numbers into a different grid. Pass `-1` and PyTorch works out that dimension for you.

```python
x = torch.arange(6)        # tensor([0, 1, 2, 3, 4, 5]), shape [6]
y = x.reshape(3, 2)        # same six numbers, now 3 rows of 2
print(y)

print(x.reshape(2, -1))    # "2 rows, you figure out the columns"
```

## 4. Doing things with tensors

Arithmetic happens elementwise — the operation runs on each number, and you get a tensor of the same shape back. Note that `*` is **not** matrix multiplication.

```python
a = torch.tensor([1, 2, 3])
b = torch.tensor([10, 20, 30])

print(a + b)   # tensor([11, 22, 33])
print(a * 2)   # tensor([2, 4, 6])
print(a * b)   # tensor([10, 40, 90]) — pairs multiplied, not "matrix math"
```

Matrix multiplication has its own operator, `@`. All the intuition you need: `@` is the core operation a neural network repeats over and over. You'll see it everywhere inside models.

```python
m = torch.tensor([[1, 2],
                  [3, 4]])
n = torch.tensor([[5, 6],
                  [7, 8]])
print(m @ n)
```

### A gentle word on broadcasting

Combine tensors of different shapes and PyTorch quietly stretches the smaller one to fit. Here the single row gets added to every row of the grid.

```python
grid = torch.ones(2, 3)        # 2x3 of ones
row  = torch.tensor([1, 2, 3]) # a single row of 3

print(grid + row)
# tensor([[2, 3, 4],
#         [2, 3, 4]])
```

## 5. Moving to the GPU

A CPU tensor and a GPU tensor behave identically — the GPU is just faster on big tensors. This `device` line lets the same notebook run with or without a GPU.

Want the GPU? **Runtime → Change runtime type → Hardware accelerator → GPU**, then run this cell again.

```python
device = "cuda" if torch.cuda.is_available() else "cpu"

x = torch.rand(2, 3)
x = x.to(device)
print(x.device)   # cuda (if a GPU is attached) or cpu otherwise
```

## 6. Talking to NumPy, and pulling out single values

Tensors and NumPy arrays are close cousins — pass data between them freely. And `.item()` pulls a single number out as a plain Python value.

```python
import numpy as np

t = torch.tensor([1.0, 2.0, 3.0])
arr = t.numpy()                 # tensor -> NumPy array
print(arr)

arr2 = np.array([4.0, 5.0, 6.0])
back = torch.from_numpy(arr2)   # NumPy array -> tensor
print(back)

score = torch.tensor(0.98)
print(score.item())             # 0.98 — a normal Python float, not a tensor
```

That's the data structure the whole series runs on: make a tensor, ask its shape, reshape it, do arithmetic, send it to a GPU, trade it with NumPy. No calculus required.
