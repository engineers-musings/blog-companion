## Loading the penguins

`import pandas as pd` brings in the data tool. `sns.load_dataset("penguins")` fetches a built-in dataset. `.dropna()` throws out rows with missing measurements. `.head()` shows the first 5 rows.

```python
import pandas as pd
import seaborn as sns

penguins = sns.load_dataset("penguins")
penguins = penguins.dropna()
penguins.head()
```

## How big is it?

`.shape` gives `(rows, columns)`. `len(...)` counts the rows. `.columns` lists the column names.

```python
print("Rows and columns:", penguins.shape)
print("Number of penguins:", len(penguins))
print("The columns are:", list(penguins.columns))
```

## Peeking at the data

A few ways to look: the first rows, the last rows, and some random rows.

```python
penguins.head(10)    # first 10 rows
```

```python
penguins.tail()      # last 5 rows
```

```python
penguins.sample(5)   # 5 random rows
```

## A quick summary

pandas can summarize all the number columns at once — count, average (`mean`), smallest (`min`), and largest (`max`).

```python
penguins.describe()
```
