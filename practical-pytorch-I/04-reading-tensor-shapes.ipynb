Reading tensor shapes without panicking. Print shapes at every boundary and label each dimension.

## A single thing vs a batch of things

The first dimension is usually the batch.

```python
import torch

one = torch.rand(10)
batch = one.unsqueeze(0)
print(one.shape)    # torch.Size([10])
print(batch.shape)  # torch.Size([1, 10]) — one example, ten features
```

## Images: [batch, channels, height, width]

Most PyTorch vision models are channels-first.

```python
img = torch.rand(1, 3, 224, 224)   # 1 image, 3 channels, 224x224
print(img.shape)
# [3, 224, 224] is one image WITHOUT the batch dim; add it with .unsqueeze(0)
```

## Classifiers: [batch, classes], and softmax on the last axis

```python
logits = torch.rand(1, 5)            # 1 example, 5 class scores
probs = logits.softmax(dim=-1)       # normalize across classes (last axis)
print(probs.shape)                   # torch.Size([1, 5])
print(probs.sum())                   # ~1.0
```

## Fixing a shape once you've read it

Read the shape, then bend it: `unsqueeze`/`squeeze` for the batch dim, `permute` for channels-last to channels-first, `reshape` to flatten. A *type* mismatch (`Float` vs `Long`) is a `.float()`/`.long()` away, not a reshape.

```python
x = torch.rand(3, 224, 224)
print(x.unsqueeze(0).shape)    # add a batch dim:  [1, 3, 224, 224]
print(x.reshape(1, -1).shape)  # flatten to a row: [1, 150528]

img = torch.rand(224, 224, 3)          # channels-last (PIL/NumPy style)
print(img.permute(2, 0, 1).shape)      # -> [3, 224, 224] channels-first
```
