## 2. Load DistilBERT and read it

`AutoModel.from_pretrained` downloads the architecture *and* its trained weights (once, then cached). Then do the same thing you did with ResNet — just `print` it and read the shape of the tree.

```python
from transformers import AutoModel

model = AutoModel.from_pretrained("distilbert-base-uncased")
print(model)
```

Three things, top to bottom: `embeddings` turn token IDs into 768-number vectors (the `Embedding(30522, 768)` table is the vocabulary), then `transformer` holds `6 x TransformerBlock` — and each block is exactly Part 13's anatomy: an `attention`, an `ffn` (two `Linear`s), and `LayerNorm`s.

## 3. Count the parameters

```python
total = sum(p.numel() for p in model.parameters())
print(f"{total:,} parameters")   # 66,362,880 parameters
```

## 4. Where the numbers live

Peek at the first few named parameters — the names mirror the tree you just printed.

```python
for name, p in list(model.named_parameters())[:4]:
    print(name, tuple(p.shape))
```

The word-embeddings table alone — 30,522 × 768 — is over 23 million numbers, more than a third of the whole model. A huge share of a language model is just the learned dictionary.

Note: `AutoModel` gives you the backbone (hidden states), not a prediction. For a task head, use an `AutoModelFor...` class (Part 17).
