## A dataset of handwriting

scikit-learn comes with little images of handwritten digits (0–9). Each one is just 8 by 8 dots.

```python
from sklearn.datasets import load_digits

digits = load_digits()
print("Number of images:", len(digits.images))
print("Each image is this big:", digits.images[0].shape)
```

## Peek inside one image

The secret: each image is a grid of numbers. Each number is how **bright** one pixel is — `0` is black, higher is brighter.

```python
print(digits.images[0])
```

## Now see it as a picture

`matplotlib` turns the number grid back into an image. The title shows its **label** — the correct answer.

```python
import matplotlib.pyplot as plt

plt.imshow(digits.images[0], cmap="gray")
plt.title("This image is labeled: " + str(digits.target[0]))
plt.show()
```
