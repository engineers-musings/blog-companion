The famous little models your toy descends from — the perceptron, linear/logistic regression, the MLP that cracked XOR, and LeNet — each a few lines you can build and `print`. No training, just the shapes.

## 1. The perceptron (1958)

Rosenblatt's perceptron is a single weighted sum plus a threshold: fire 1 if the sum clears 0, else 0. That's a single `nn.Linear` with a threshold on the end. One layer like this famously *cannot* learn XOR.

```python
import torch
import torch.nn as nn

perceptron = nn.Linear(2, 1)        # two inputs, one output — Rosenblatt's unit
score = perceptron(torch.tensor([[1.0, 1.0]]))
print(score)                        # a weighted sum; the perceptron fires 1 if it's > 0
```

## 2. Linear and logistic regression

Drop the hard threshold and the same one layer is a workhorse. A bare `nn.Linear` *is* linear regression; add a sigmoid and it's logistic regression — the most widely used classifier on the planet.

```python
linreg = nn.Linear(4, 1)            # 4 features in, one number out — linear regression

logreg = nn.Sequential(
    nn.Linear(4, 1),                # weighted sum
    nn.Sigmoid(),                   # squash to a probability between 0 and 1
)
print(logreg(torch.rand(1, 4)))     # e.g. tensor([[0.62]]) — a probability
```

## 3. The MLP that cracked XOR

Stack a hidden layer with a nonlinearity between input and output and the wall comes down — this is a multi-layer perceptron, the shape you built last post. Once trained (Phase II), it solves the XOR the perceptron couldn't. The Universal Approximation Theorem says one hidden layer, wide enough, can approximate *any* continuous function.

```python
xor_net = nn.Sequential(
    nn.Linear(2, 2),                # a hidden layer of 2 units
    nn.ReLU(),                      # the nonlinear kink that makes depth count
    nn.Linear(2, 1),
)
print(xor_net)
```

Perceptron → logistic regression → MLP: each step is the previous one plus a little depth, all built from pieces you already know. But every one treats its input as a flat list of numbers — fine for a row of features, wrong for an image, where which pixel sits next to which is the point.
