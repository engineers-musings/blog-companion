How a model picks the next word: greedy vs sampling, temperature, top-k, top-p, repetition penalty.

```python
!pip install -q transformers
```

```python
from transformers import pipeline
gen = pipeline("text-generation", model="distilgpt2")
```

## Greedy: always the top token (tends to loop)

```python
print(gen("My favorite hobby is", max_new_tokens=40, do_sample=False)[0]["generated_text"])
```

## Sampling with the knobs

`temperature` = wildness, `top_p` = nucleus pool, `repetition_penalty` = stop the broken record. (These only apply with `do_sample=True`.)

```python
print(gen(
    "Once upon a time,",
    max_new_tokens=60,
    do_sample=True,
    temperature=0.7,
    top_p=0.9,
    repetition_penalty=1.15,
)[0]["generated_text"])
```

Try `temperature=0.2` (focused) vs `temperature=1.2` (wild). For a reproducible run, `from transformers import set_seed; set_seed(0)`.
