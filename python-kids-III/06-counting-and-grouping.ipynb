## Load the data first

```python
import pandas as pd
import seaborn as sns
penguins = sns.load_dataset("penguins").dropna()
```

## How many of each? value_counts

`.value_counts()` counts how many times each value appears in a column.

```python
penguins["species"].value_counts()
```

```python
penguins["island"].value_counts()
```

## Average per group: groupby

`groupby` splits the data into groups, then computes something for each group. Read it left to right: take `penguins`, **group by** `species`, look at `body_mass_g`, and find the `mean` of each group.

```python
penguins.groupby("species")["body_mass_g"].mean()
```

## Other things per group

Swap `.mean()` for `.max()` (biggest in each group), `.min()` (smallest), or `.count()` (how many in each group).

```python
penguins.groupby("species")["body_mass_g"].max()
```

```python
penguins.groupby("island")["species"].count()
```
