## Setup: train the model from last lesson

Start by training the digit recognizer from the previous lesson.

```python
from sklearn.datasets import load_digits
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
digits = load_digits()
X, y = digits.data, digits.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
model = KNeighborsClassifier()
model.fit(X_train, y_train)
```

## Watch it predict

Label the test images, then look at a few — the picture, what the AI guessed, and the true answer. `.reshape(8, 8)` folds the 64 numbers back into a picture.

```python
predictions = model.predict(X_test)

for i in range(5):
    plt.imshow(X_test[i].reshape(8, 8), cmap="gray")
    plt.title("AI guessed " + str(predictions[i]) + " — really " + str(y_test[i]))
    plt.show()
```

## Hunt for the mistakes

Find the ones it got wrong and look closely — they're usually genuinely messy, confusing digits.

```python
for i in range(len(X_test)):
    if predictions[i] != y_test[i]:
        plt.imshow(X_test[i].reshape(8, 8), cmap="gray")
        plt.title("AI said " + str(predictions[i]) + ", but it's " + str(y_test[i]))
        plt.show()
```

### Try it 🎯

Count how many it got wrong out of the test set.

```python
wrong = 0
for i in range(len(X_test)):
    if predictions[i] != y_test[i]:
        wrong = wrong + 1
print("Got", wrong, "wrong out of", len(X_test))
```

## Try to do better

Two simple ways to push accuracy up. First, a different number of neighbors.

```python
# 1. Try a different number of neighbors
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
print("With 3 neighbors:", model.score(X_test, y_test))
```

Second, give it MORE training data (a smaller test set).

```python
# 2. Give it MORE training data (smaller test set)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
model.fit(X_train, y_train)
print("With more training data:", model.score(X_test, y_test))
```
